```python
from sklearn.linear_model import LinearRegression
```

```python
LR = LinearRegression()
# train
LR.fit(x,y)

# predict with trai data
y_pred =LR.predict(x)
```

```python
plt.plot(x,y,'o')
```

```python
print(LR.coef_)
print(LR.intercept_)
```

```python
# create user defined function with name mse
# inputs y_pred ,y
# output is mean squared error
def mse(y_pred,y):
    return ((y_pred-y)**2).mean()
```
