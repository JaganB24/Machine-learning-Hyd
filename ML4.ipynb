Create a new array<br>
 - a>=50 and a<=150 - Good<br>
 - a<50 - Bad<br>
 - a>150 - Excellent<br>

```python
def number_cat(n):
    if n<50:
        return 'Bad'
    elif(n>=50) and (n<=150):
        return 'Good'
    else:
        return'Excellent'
```

```python
number_cat(20)
```

```python
number_cat_vect = np.vectorize(number_cat)
```

```python
number_cat_vect(a)
```

```python
import pandas as pd
cat = pd.Series(number_cat_vect(a))
cat
```

```python
cat.value_counts()
```

```python
y_values = cat.value_counts()
x_values = cat.value_counts().index
plt.bar(x_values,y_values)
```

```python
def distance(p1,p2):
    d=math.sqrt((p1[0]-p2[0])**2+(p1[1]-p2[1])**2)
    return d
p1 = [1,2]
p2 = [2,3]
ab =distance(p1,p2)
abs(ab)
```

```python
def absd(p1,p2):
    d=math.sqrt(abs((p1[0]-p2[0])**2+(p1[1]-p2[1])**2))
    return d
p1 = [-1,-2]
p2 = [-2,-3]
absd(p1,p2)
```
