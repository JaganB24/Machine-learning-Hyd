```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv(('https://raw.githubusercontent.com/sumathi16/Machine-Learning-with-Python/master/data/global_sales_data/master.csv'))
```

```python
data.head()
```

```python
data.drop('Unnamed: 0' ,axis =1,inplace=True)
```

```python
sns.distplot(data['Sales'])
```

```python
sns.distplot(data['Order_Quantity'])
```

```python
sns.distplot(data['Order_Quantity'],hist=False)
```

```python
sns.distplot(data['Order_Quantity'],hist=False,rug=True)
```
